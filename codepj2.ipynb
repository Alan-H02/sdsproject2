```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")

from IPython.display import display

from scipy import sparse
from scipy.stats import ks_2samp, spearmanr, kendalltau

from sklearn.model_selection import train_test_split, RepeatedStratifiedKFold
from sklearn.preprocessing import OneHotEncoder, RobustScaler, FunctionTransformer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.utils.class_weight import compute_class_weight
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    roc_auc_score, average_precision_score, precision_recall_curve,
    f1_score, precision_score, recall_score, accuracy_score,
    brier_score_loss, roc_curve
)
from sklearn.calibration import calibration_curve
from sklearn.neighbors import LocalOutlierFactor, NearestNeighbors
from sklearn.cluster import DBSCAN
from sklearn.utils import check_random_state

import jax
import jax.numpy as jnp
import jax.random as random
import numpyro
import numpyro.distributions as dist
import numpyro.optim as optim
from numpyro.infer import SVI, Trace_ELBO, Predictive
from numpyro.infer.autoguide import AutoLowRankMultivariateNormal as AutoLowRank

import torch
from torch import nn
from torch.utils.data import TensorDataset, DataLoader

import shap
try:
    from xgboost import XGBClassifier
except ImportError:
    XGBClassifier = None

def base_of(col: str) -> str:
    """Convert engineered names back to base features, e.g. 'num__dur'->'dur', 'bin__x'->'x', 'cat__proto_tcp'->'proto'."""
    if col.startswith(("num__", "bin__")):
        return col.split("__", 1)[1]
    if col.startswith("cat__"):
        return col.split("__", 1)[1].split("_", 1)[0]
    return col

RANDOM_STATE = 25
pd.set_option("display.max_columns", 200)

df_train = pd.read_csv("UNSW_NB15_training-set.csv")
df_test  = pd.read_csv("UNSW_NB15_testing-set.csv")

print("Train shape:", df_train.shape)
print("Test shape:", df_test.shape)
print("Column names match:", set(df_train.columns) == set(df_test.columns))
print("Columns:", df_train.columns.tolist())

def label_summary(df: pd.DataFrame, split_name: str) -> pd.DataFrame:
    vc = df["label"].value_counts(dropna=False).sort_index()
    prop = vc / vc.sum()
    return pd.DataFrame({
        "split": split_name,
        "label": vc.index,
        "count": vc.values,
        "ratio": prop.values
    })

lbl_sum = pd.concat([
    label_summary(df_train, "Train"),
    label_summary(df_test,  "Test")
], ignore_index=True)

print("\n[Label distribution (table)]")
display(lbl_sum)

pivot_ratio = lbl_sum.pivot(index="split", columns="label", values="ratio").fillna(0)
pivot_ratio = pivot_ratio.reindex(columns=sorted(pivot_ratio.columns))

ax = pivot_ratio.plot(kind="bar", stacked=True, figsize=(6,4))
ax.set_ylim(0, 1)
ax.set_ylabel("Proportion")
ax.set_title("Label Distribution by Split")
ax.legend([f"label={c}" for c in pivot_ratio.columns], loc="upper right")
for container in ax.containers:
    ax.bar_label(container, fmt="%f", label_type="center")

plt.tight_layout(); plt.show()

def clean_bad_rows(train: pd.DataFrame, test: pd.DataFrame, tol_abs: float = 1e-6, verbose: bool = True):
    """
    Minimal rules:
      - Any NaN in a row -> drop
      - TCP timing consistency: |tcprtt - (synack + ackdat)| > tol_abs -> drop
      - Non-negativity: specified columns must be >= 0
    """
    def has_nan(df: pd.DataFrame) -> pd.Series:
        return df.isna().any(axis=1)

    def tcp_inconsistent(df: pd.DataFrame, tol=1e-6) -> pd.Series:
        cols = ['tcprtt','synack','ackdat']
        valid = df[cols].notna().all(axis=1)
        diff = (df.loc[valid, 'tcprtt'] - (df.loc[valid, 'synack'] + df.loc[valid, 'ackdat'])).abs()
        out = pd.Series(False, index=df.index)
        out.loc[valid] = diff > tol
        return out

    nonneg_cols = ['dur','spkts','dpkts','sbytes','dbytes','sload','dload',
        'smean','dmean','sinpkt','dinpkt','sjit','djit','tcprtt','synack','ackdat','rate']

    def neg_masks(df: pd.DataFrame) -> dict:
        return {f'neg_{c}': (df[c].notna() & (df[c] < 0)) for c in nonneg_cols}

    if verbose:
        print("\n[Missing values by column]")
        for name, df in [("Train", train), ("Test", test)]:
            miss = df.isna().sum()
            miss = miss[miss > 0].sort_values(ascending=False)
            if len(miss) == 0:
                print(f"  - {name}: no missing values")
            else:
                print(f"  - {name}: missing values (by column)")
                display(miss.to_frame("n_missing"))

    def collect_and_report(df: pd.DataFrame, name: str) -> tuple[dict, int]:
        masks = {'has_nan': has_nan(df), 'tcp_timing_inconsistent': tcp_inconsistent(df, tol_abs)}
        masks.update(neg_masks(df))
        total = 0
        if verbose:
            print(f"\n[Data quality checks] {name}:")
        for k, m in masks.items():
            cnt = int(m.sum()); total += cnt
            if verbose and cnt > 0:
                print(f"  - {k}: {cnt}")
        if verbose and total == 0:
            print("  No abnormal records found.")
        return masks, total

    tr_masks, tr_total = collect_and_report(train, "Train")
    te_masks, te_total = collect_and_report(test,  "Test")

    if tr_total == 0 and te_total == 0:
        if verbose:
            print("\n[Cleaning] No anomalies detected, skipping.")
        return train.copy(), test.copy()

    def apply_clean(df: pd.DataFrame, masks: dict, name: str) -> pd.DataFrame:
        keep = pd.Series(True, index=df.index)
        removed = {}
        for k, m in masks.items():
            bad = keep & m.fillna(False)
            n = int(bad.sum())
            if n > 0:
                removed[k] = n
                keep &= ~bad
        if verbose:
            print(f"\n[Cleaning applied] {name}: removed rows {int((~keep).sum())}")
            for k, v in removed.items():
                print(f"  - {k}: {v}")
        return df.loc[keep].copy()

    tr_clean = apply_clean(train, tr_masks, "Train")
    te_clean = apply_clean(test,  te_masks, "Test")
    if verbose:
        print("\n[Shapes after cleaning] Train:", tr_clean.shape, "  Test:", te_clean.shape)
    return tr_clean, te_clean

df_train, df_test = clean_bad_rows(df_train, df_test, tol_abs=1e-6, verbose=True)

def drop_abnormal_is_ftp_login(df: pd.DataFrame, name: str) -> pd.DataFrame:
    if 'is_ftp_login' not in df.columns:
        return df
    mask_abn = ~df['is_ftp_login'].isin([0, 1]) & df['is_ftp_login'].notna()
    n_abn = int(mask_abn.sum())
    if n_abn > 0:
        ratio = n_abn / len(df)
        print(f"[{name}] is_ftp_login abnormal values: {n_abn} rows ({ratio:.4%}), removing them.")
        df = df.loc[~mask_abn].copy()
        print(f"[{name}] shape after removal: {df.shape}")
    else:
        print(f"[{name}] is_ftp_login has no abnormal values.")
    return df

df_train = drop_abnormal_is_ftp_login(df_train, "Train")
df_test  = drop_abnormal_is_ftp_login(df_test,  "Test")

id_col         = "id"
target_col     = "label"
aux_target_col = "attack_cat"
categorical_cols = ["proto", "service", "state"]
binary_cols      = ["is_sm_ips_ports", "is_ftp_login"]
reserved = {id_col, target_col, aux_target_col, *categorical_cols, *binary_cols}
numeric_not_binary = [c for c in df_train.columns if c not in reserved]
numeric_cols = binary_cols + numeric_not_binary

print("Feature grouping")
print("  id_col           :", id_col)
print("  target_col       :", target_col)
print("  aux_target_col   :", aux_target_col)
print("  categorical_cols :", categorical_cols)
print("  binary_cols      :", binary_cols)
print("  numeric_not_bin #:", len(numeric_not_binary))
print("  numeric_cols   # :", len(numeric_cols))

def stacked_count_by_label_grouped(df, cat_col, label_col="label"):
    ct = pd.crosstab(df[cat_col], df[label_col])
    for col in [0, 1]:
        if col not in ct.columns:
            ct[col] = 0
    ct = ct[[0, 1]]

    all0_mask = (ct[1] == 0)
    all1_mask = (ct[0] == 0)
    mixed = ct[~(all0_mask | all1_mask)].copy()

    all0_cnt = ct.loc[all0_mask].sum()
    all1_cnt = ct.loc[all1_mask].sum()

    rows = []
    if int(all0_mask.sum()) > 0:
        rows.append(pd.Series({"cat": f"ALL-0 ({int(all0_mask.sum())} cats)", 0: int(all0_cnt[0]), 1: int(all0_cnt[1])}))
    for k, r in mixed.iterrows():
        rows.append(pd.Series({"cat": str(k), 0: int(r[0]), 1: int(r[1])}))
    if int(all1_mask.sum()) > 0:
        rows.append(pd.Series({"cat": f"ALL-1 ({int(all1_mask.sum())} cats)", 0: int(all1_cnt[0]), 1: int(all1_cnt[1])}))

    return pd.DataFrame(rows)

fig, axes = plt.subplots(1, 3, figsize=(20, 7), sharex=False)
for i, cat in enumerate(categorical_cols):
    plot_df = stacked_count_by_label_grouped(df_train, cat, target_col)
    ax = axes[i]
    y = np.arange(len(plot_df))
    ax.barh(y, plot_df[0], label="Normal (0)")
    ax.barh(y, plot_df[1], left=plot_df[0], label="Attack (1)")
    ax.set_yticks(y)
    ax.set_yticklabels(plot_df["cat"], fontsize=15)
    ax.set_xlabel("Count", fontsize=13)
    ax.set_title(f"{cat} × {target_col}", fontsize=18)
    totals = (plot_df[0] + plot_df[1]).values
    for j, (c0, c1, tot) in enumerate(zip(plot_df[0].values, plot_df[1].values, totals)):
        ax.text(tot + max(50, tot*0.003), j, f"{c0}:{c1}", va="center", ha="left", fontsize=13)
    ax.set_xlim(0, totals.max() * 1.12)
handles, labels = axes[0].get_legend_handles_labels()
fig.legend(handles, labels, title="Label", loc="upper center", ncol=2, fontsize=13)
plt.tight_layout(rect=[0,0,1,0.92])
plt.show()

key_num_cols = [
    "dur","spkts","dpkts","sbytes","dbytes","rate","sload","dload",
    "sttl","dttl","synack","ackdat","tcprtt","sinpkt","dinpkt"
]
print("\n[Numeric features] Key histograms (log1p, 25% stratified sample)")

df_plot = df_train.groupby(target_col, group_keys=False).apply(
    lambda x: x.sample(frac=0.25, random_state=RANDOM_STATE)
).reset_index(drop=True)

n = len(key_num_cols)
ncols = 3
nrows = (n + ncols - 1) // ncols
plt.figure(figsize=(4*ncols, 3*nrows))
for i, c in enumerate(key_num_cols, 1):
    plt.subplot(nrows, ncols, i)
    sns.histplot(np.log1p(df_plot[c]), bins=50, kde=False)
    plt.title(f"{c} (log1p)")
plt.tight_layout()
plt.show()

def plot_group_violin(df, feats, label_col="label"):
    n = len(feats)
    ncols = 3
    nrows = (n + ncols - 1) // ncols
    fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(5*ncols, 3.8*nrows))
    axes = np.array(axes).reshape(-1)
    for i, f in enumerate(feats):
        tmp = df[[f, label_col]].copy()
        tmp["val"] = np.log1p(tmp[f])
        ax = axes[i]
        sns.violinplot(data=tmp, x=label_col, y="val", cut=0, inner="quartile", ax=ax)
        ax.set_title(f"{f} (log1p)")
        ax.set_xlabel(label_col); ax.set_ylabel("")
    plt.tight_layout()
    plt.show()

print("\n[Numeric features] Violin plots by label (log1p, same sampling)")
plot_group_violin(df_plot, key_num_cols, label_col=target_col)

print("\n[Binary features] 0/1 counts by label")
for b in binary_cols:
    tab = pd.crosstab(df_train[target_col], df_train[b], dropna=False)
    for v in [0, 1]:
        if v not in tab.columns:
            tab[v] = 0
    tab = tab[[0, 1]]
    tab_prop = tab.div(tab.sum(axis=1), axis=0)

    print(f"\n[{b}] counts")
    display(tab)
    print(f"[{b}] proportions")
    display(tab_prop.style.format("{:.3f}"))

def ks_subsample(train, test, num_cols, n_per_split=10000, n_repeats=50, random_state=42):
    results = []
    rng = np.random.default_rng(random_state)
    for c in num_cols:
        if c not in train.columns or c not in test.columns:
            continue
        a = pd.to_numeric(train[c], errors="coerce").dropna().values
        b = pd.to_numeric(test[c], errors="coerce").dropna().values
        if len(a) == 0 or len(b) == 0:
            continue
        stat_full, p_full = ks_2samp(a, b)
        subs = []
        for _ in range(n_repeats):
            sub_a = rng.choice(a, size=min(len(a), n_per_split), replace=False)
            sub_b = rng.choice(b, size=min(len(b), n_per_split), replace=False)
            s, _ = ks_2samp(sub_a, sub_b)
            subs.append(s)
        results.append({
            "feature": c,
            "ks_full_stat": stat_full,
            "ks_full_p": p_full,
            "ks_sub_mean": np.mean(subs),
            "train_mean": np.mean(a),
            "test_mean": np.mean(b)
        })
    return pd.DataFrame(results).sort_values("ks_sub_mean", ascending=False).reset_index(drop=True)

ks_result = ks_subsample(
    df_train, 
    df_test, 
    num_cols=numeric_not_binary, 
    n_per_split=10000, 
    n_repeats=50
)
print("KS subsample results (sorted by mean desc):")
display(ks_result.head(20))

drop_cols   = [id_col, aux_target_col]
train_df    = df_train.drop(columns=drop_cols).copy()
test_df     = df_test.drop(columns=drop_cols).copy()

y_train_full = train_df[target_col].astype(int)
X_train_full = train_df.drop(columns=[target_col])
y_test       = test_df[target_col].astype(int)
X_test       = test_df.drop(columns=[target_col])

X_tr, X_val, y_tr, y_val = train_test_split(
    X_train_full, y_train_full, test_size=0.20,
    stratify=y_train_full, random_state=RANDOM_STATE
)

val_index  = X_val.index.copy()
test_index = X_test.index.copy()

ZERO_FLAG_MIN_PROP = 0.80
zero_flag_cols = []
for c in numeric_not_binary:
    if (X_tr[c] == 0).mean() >= ZERO_FLAG_MIN_PROP:
        flag = f"{c}_is_zero"
        X_tr[flag]   = (X_tr[c] == 0).astype(int)
        X_val[flag]  = (X_val[c] == 0).astype(int)
        X_test[flag] = (X_test[c] == 0).astype(int)
        zero_flag_cols.append(flag)

binary_cols_ext = binary_cols + zero_flag_cols

def _log1p_safe(x: pd.DataFrame) -> pd.DataFrame:
    return np.log1p(x.clip(lower=-1.0))

log1p_tf = FunctionTransformer(_log1p_safe, feature_names_out="one-to-one")
robust   = RobustScaler(with_centering=True, with_scaling=True, quantile_range=(25.0, 75.0))

try:
    ohe = OneHotEncoder(handle_unknown="ignore", sparse_output=True)
except TypeError:
    ohe = OneHotEncoder(handle_unknown="ignore", sparse=True)

oh_preprocessor = ColumnTransformer(
    transformers=[
        ("num", Pipeline([("log1p", log1p_tf), ("robust", robust)]), numeric_not_binary),
        ("bin", "passthrough", binary_cols_ext),
        ("cat", ohe, categorical_cols),
    ],
    remainder="drop",
    sparse_threshold=1.0
)

oh_preprocessor.fit(X_tr)

X_tr_oh  = oh_preprocessor.transform(X_tr)
X_val_oh = oh_preprocessor.transform(X_val)
X_tst_oh = oh_preprocessor.transform(X_test)

try:
    oh_feature_names = oh_preprocessor.get_feature_names_out()
except Exception:
    oh_feature_names = np.array([f"f{i}" for i in range(X_tr_oh.shape[1])])

print("[One-Hot] shapes:", X_tr_oh.shape, X_val_oh.shape, X_tst_oh.shape,
      "| sparse =", sparse.issparse(X_tr_oh))
print("New zero-indicator columns:", zero_flag_cols)

classes           = np.unique(y_tr)
class_weights     = compute_class_weight(class_weight="balanced", classes=classes, y=y_tr)
class_weight_dict = {int(c): float(w) for c, w in zip(classes, class_weights)}
print("[Class weights]", class_weight_dict)

N_SPLITS, N_REPEATS = 5, 3
FINAL_K             = 20
K_FOR_STAB          = 30

feat_names = np.array(oh_feature_names)
if getattr(feat_names, "ndim", 1) == 0:
    feat_names = np.array([str(feat_names)])

base_names     = np.array([base_of(c) for c in feat_names])
all_base_feats = sorted(set(base_names))

rskf = RepeatedStratifiedKFold(n_splits=N_SPLITS, n_repeats=N_REPEATS, random_state=RANDOM_STATE)
sum_imp, sum_sqimp = pd.Series(0.0, index=all_base_feats), pd.Series(0.0, index=all_base_feats)
fold_rankings, n_models = [], 0

for fold_id, (idx_tr, idx_va) in enumerate(rskf.split(X_tr_oh, y_tr), 1):
    rf = RandomForestClassifier(
        n_estimators=100, max_features="sqrt",
        class_weight=class_weight_dict, n_jobs=-1,
        random_state=RANDOM_STATE + fold_id
    )
    rf.fit(X_tr_oh[idx_tr], y_tr.iloc[idx_tr])

    imp_raw = pd.DataFrame({"base": base_names, "imp": rf.feature_importances_})
    imp_agg = imp_raw.groupby("base")["imp"].sum().reindex(all_base_feats).fillna(0.0)

    sum_imp   += imp_agg
    sum_sqimp += imp_agg**2
    n_models  += 1

    fold_rankings.append(imp_agg.sort_values(ascending=False).index.tolist())

imp_mean = sum_imp / n_models
imp_std  = np.sqrt(np.clip((sum_sqimp / n_models) - imp_mean**2, 0, None))
imp_summary = (
    pd.DataFrame({"feature": imp_mean.index, "importance_mean": imp_mean, "importance_std": imp_std})
    .sort_values("importance_mean", ascending=False)
    .reset_index(drop=True)
)
print("[RF] CV aggregated importance (Top-25 preview):")
display(imp_summary.head(25))

def ranking_to_rankvec(ranking_list, universe):
    pos = {f: i for i, f in enumerate(ranking_list)}
    return np.array([pos.get(f, len(universe)) for f in universe], dtype=float)

rho_vals, tau_vals, jacc_vals = [], [], []
for i in range(len(fold_rankings)):
    for j in range(i+1, len(fold_rankings)):
        r1, r2 = fold_rankings[i], fold_rankings[j]
        v1, v2 = ranking_to_rankvec(r1, all_base_feats), ranking_to_rankvec(r2, all_base_feats)
        rho_vals.append(spearmanr(v1, v2).correlation)
        tau_vals.append(kendalltau(v1, v2).correlation)
        top1, top2 = set(r1[:K_FOR_STAB]), set(r2[:K_FOR_STAB])
        jacc_vals.append(len(top1 & top2) / max(len(top1 | top2), 1))

stab = {
    "Spearman(rho)_mean": float(np.nanmean(rho_vals)),
    "Spearman(rho)_min":  float(np.nanmin(rho_vals)),
    "Kendall(tau)_mean":  float(np.nanmean(tau_vals)),
    "Kendall(tau)_min":   float(np.nanmin(tau_vals)),
    "TopK_Jaccard_mean":  float(np.nanmean(jacc_vals)),
    "TopK_Jaccard_min":   float(np.nanmin(jacc_vals)),
    "pairs":              int(len(rho_vals)),
}
print("\n[Stability] CV ranking stability (report only)")
print(stab)

selected = imp_summary["feature"].head(FINAL_K).tolist()
print(f"\n[Selection] Take stable Top-{FINAL_K} base features directly:")
print(selected)

mask_selK   = np.array([b in set(selected) for b in base_names])
X_tr_oh_sel  = X_tr_oh[:, mask_selK]
X_val_oh_sel = X_val_oh[:, mask_selK]

rf_sel = RandomForestClassifier(
    n_estimators=300, max_features="sqrt",
    class_weight=class_weight_dict, n_jobs=-1, random_state=RANDOM_STATE
).fit(X_tr_oh_sel, y_tr)

proba_sel = rf_sel.predict_proba(X_val_oh_sel)[:, 1]
pred_sel  = (proba_sel >= 0.5).astype(int)

def _eval_overall(y_true, proba, pred, name):
    metrics = {
        "ROC-AUC": roc_auc_score(y_true, proba),
        "PR-AUC":  average_precision_score(y_true, proba),
        "F1":      f1_score(y_true, pred),
        "Prec":    precision_score(y_true, pred, zero_division=0),
        "Recall":  recall_score(y_true, pred, zero_division=0),
        "Acc":     float((pred == y_true).mean()),
    }
    print(f"[{name}] Overall:", metrics)
    return metrics

_ = _eval_overall(y_val.values, proba_sel, pred_sel, f"RF-OHE-Selected{FINAL_K}")

if ("attack_cat" in df_train.columns) and (len(val_index) > 0):
    val_attack_cat = df_train.loc[val_index, "attack_cat"].astype(str)
    rows = []
    for c in sorted(val_attack_cat.dropna().unique()):
        m = (val_attack_cat == c)
        if m.sum() == 0: 
            continue
        rows.append({
            "attack_cat": c, "N": int(m.sum()),
            "Recall(attack)": float((pred_sel[m] == 1).mean()),
            "AvgProb": float(np.mean(proba_sel[m]))
        })
    df_res = pd.DataFrame(rows).sort_values("Recall(attack)", ascending=False)
    print(f"\n[RF-OHE-Selected{FINAL_K}] Per-attack detection recall (Validation)")
    display(df_res)

print(f"\nDim info: Selected{FINAL_K} → Val columns = {X_val_oh_sel.shape[1]}")

df_plot = (
    imp_summary[imp_summary["feature"].isin(selected)]
    .sort_values("importance_mean", ascending=True)
)

plt.figure(figsize=(8, 6))
bars = plt.barh(
    y=df_plot["feature"],
    width=df_plot["importance_mean"],
    xerr=df_plot["importance_std"],
    capsize=3, alpha=0.9, edgecolor="#333", linewidth=0.6
)
xmax = df_plot["importance_mean"].max()
for rect, mean in zip(bars, df_plot["importance_mean"]):
    x = rect.get_width()
    y = rect.get_y() + rect.get_height() / 2
    plt.text(x + 0.01 * xmax, y, f"{mean:.3f}", va="center", fontsize=10)

plt.title(f"Selected Top-{FINAL_K} Features — RF Importance (CV mean)")
plt.xlabel("Importance")
plt.ylabel("Feature")
plt.tight_layout()
plt.show()

def to_dense32(X):
    return (X.toarray() if sparse.issparse(X) else np.asarray(X)).astype(np.float32)

keep_mask  = np.array([b in set(selected) for b in base_names])
X_tr_oh_red  = X_tr_oh[:, keep_mask]
X_val_oh_red = X_val_oh[:, keep_mask]
X_tst_oh_red = X_tst_oh[:, keep_mask]

Xtr, Xval, Xtst = map(to_dense32, [X_tr_oh_red, X_val_oh_red, X_tst_oh_red])
ytr  = np.asarray(y_tr).astype(np.int32)
yval = np.asarray(y_val).astype(np.int32)
ytes = np.asarray(y_test).astype(np.int32)

print("[OHE] shapes:", Xtr.shape, Xval.shape, Xtst.shape,
      "| kept base features:", len(set(selected)))

cw_dict = class_weight_dict
print("[Class Weights]", cw_dict)
w0 = jnp.asarray(cw_dict.get(0, 1.0), dtype=jnp.float32)
w1 = jnp.asarray(cw_dict.get(1, 1.0), dtype=jnp.float32)

def model_uncertain(x, y=None, hidden_dim=64, class_weight=None):
    n, m = x.shape
    prec_nn = numpyro.sample("prec_nn", dist.Gamma(20., 20.))
    sd1 = 1.0 / jnp.sqrt(prec_nn * (m + 1))
    sd2 = 1.0 / jnp.sqrt(prec_nn * (hidden_dim + 1))

    with numpyro.plate("l1_hidden", hidden_dim, dim=-1):
        b1 = numpyro.sample("b1", dist.Normal(0., sd1))
        with numpyro.plate("l1_feat", m, dim=-2):
            w1_ = numpyro.sample("w1", dist.Normal(0., sd1))

    with numpyro.plate("l2_hidden", hidden_dim, dim=-1):
        w2 = numpyro.sample("w2", dist.Normal(0., sd2))
    b2 = numpyro.sample("b2", dist.Normal(0., sd2))

    tau = numpyro.sample("tau", dist.LogNormal(0.0, 0.5))
    h     = jnp.maximum(x @ w1_ + b1, 0.)
    logit = (h @ w2) + b2
    logit_scaled = logit / tau
    prob  = jax.nn.sigmoid(logit_scaled)
    numpyro.deterministic("p", prob)

    if y is not None:
        with numpyro.plate("data", n):
            ll = dist.Bernoulli(logits=logit_scaled).log_prob(y)
            if class_weight is None:
                w = jnp.ones_like(ll)
            else:
                w = jnp.where(y == 1, w1, w0).astype(jnp.float32)
            numpyro.factor("weighted_loglik", jnp.sum(w * ll))

def do_SVI_uncertain(model_fn, rng_key, x, y, hidden_dim=64, steps=3000, lr=1e-2, lowrank_rank=50, class_weight=None):
    guide = AutoLowRank(model_fn, rank=min(lowrank_rank,  max(10, hidden_dim)))
    print(f"[Guide] AutoLowRankMultivariateNormal (rank={min(lowrank_rank, max(10, hidden_dim))})")
    svi   = SVI(model_fn, guide, optim.Adam(lr), Trace_ELBO())
    results = svi.run(
        rng_key=rng_key, num_steps=steps,
        x=jnp.array(x), y=jnp.array(y),
        hidden_dim=hidden_dim, class_weight=class_weight
    )
    return results.losses, results.params, guide

def posterior_mean_var_prob(pred_obj):
    p_draws = np.asarray(pred_obj["p"])
    p_mean  = p_draws.mean(axis=0)
    p_var   = p_draws.var(axis=0)
    return p_mean, p_var

def best_f1_threshold(y, proba):
    ps, rs, ths = precision_recall_curve(y, proba)
    f1s = 2*ps*rs/(ps+rs+1e-12)
    idx = np.argmax(f1s[:-1])
    return float(ths[idx])

def best_bacc_threshold(y, proba):
    fpr, tpr, ths = roc_curve(y, proba)
    if np.isinf(ths[0]):
        fpr, tpr, ths = fpr[1:], tpr[1:], ths[1:]
    bacc = (tpr + (1.0 - fpr)) / 2.0
    idx  = int(np.argmax(bacc))
    return float(ths[idx]), float(bacc[idx])

def eval_block(name, y, proba, thr):
    pred = (proba >= thr).astype(int)
    out = {
        "ROC-AUC": roc_auc_score(y, proba),
        "PR-AUC":  average_precision_score(y, proba),
        "Acc":     float((pred==y).mean()),
        "F1":      f1_score(y, pred),
        "Precision": precision_score(y, pred, zero_division=0),
        "Recall":    recall_score(y, pred, zero_division=0),
        "Brier":     brier_score_loss(y, proba),
    }
    print(f"[{name}]", out)
    return pred, out

numpyro.set_platform("cpu")
rng_key, key_val_pred, key_tst_pred, key_tr_pred = random.split(random.PRNGKey(42), 4)

H, steps, nsamples = 64, 3000, 1500
losses, params, guide = do_SVI_uncertain(
    model_uncertain, rng_key, Xtr, ytr,
    hidden_dim=H, steps=steps, lr=1e-2, lowrank_rank=64, class_weight=True
)

predictor = Predictive(model=model_uncertain, guide=guide, params=params, num_samples=nsamples)

pred_trn = predictor(rng_key=key_tr_pred,  x=jnp.array(Xtr),  y=None, hidden_dim=H, class_weight=None)
pred_val = predictor(rng_key=key_val_pred, x=jnp.array(Xval), y=None, hidden_dim=H, class_weight=None)
pred_tst = predictor(rng_key=key_tst_pred, x=jnp.array(Xtst), y=None, hidden_dim=H, class_weight=None)

p_trn, u_trn = posterior_mean_var_prob(pred_trn)
p_val, u_val = posterior_mean_var_prob(pred_val)
p_tst, u_tst = posterior_mean_var_prob(pred_tst)

plt.figure(figsize=(6,3.6))
plt.plot(np.asarray(losses))
plt.xlabel("Step")
plt.ylabel("ELBO (negative)")
plt.title("SVI Loss Curve")
plt.tight_layout()
plt.show()

thr_f1 = best_f1_threshold(yval, p_val)
print(f"[threshold] BNN best-F1 on Val: {thr_f1:.4f}")

thr_bacc, bacc_val = best_bacc_threshold(yval, p_val)
print(f"[threshold] BNN best-BAcc on Val: thr={thr_bacc:.4f}, BAcc={bacc_val:.4f}")

_ = eval_block("VAL@BNN(F1-thr)",   yval, p_val, thr_f1)
_ = eval_block("TEST@BNN(F1-thr)",  ytes, p_tst, thr_f1)
_ = eval_block("VAL@BNN(BAcc-thr)",  yval, p_val, thr_bacc)
_ = eval_block("TEST@BNN(BAcc-thr)", ytes, p_tst, thr_bacc)

def plot_roc_pr_single(y, proba, label="BNN", prefix="Val"):
    fpr, tpr, _ = roc_curve(y, proba)
    ps, rs, _   = precision_recall_curve(y, proba)
    auc_roc = roc_auc_score(y, proba)
    ap_pr   = average_precision_score(y, proba)

    plt.figure(figsize=(11,4))
    plt.subplot(1,2,1)
    plt.plot(fpr, tpr, label=f"{label} (AUC={auc_roc:.3f})")
    plt.plot([0,1],[0,1],'--',alpha=.5)
    plt.xlabel("FPR"); plt.ylabel("TPR")
    plt.title(f"{prefix} ROC — {label}")
    plt.legend()

    plt.subplot(1,2,2)
    plt.plot(rs, ps, label=f"{label} (AP={ap_pr:.3f})")
    plt.xlabel("Recall"); plt.ylabel("Precision")
    plt.title(f"{prefix} PR — {label}")
    plt.legend()

    plt.tight_layout()
    plt.show()

def reliability_plot(y, proba, title):
    pt, pp = calibration_curve(y, proba, n_bins=10, strategy="quantile")
    ece = float(np.mean(np.abs(pt - pp)))
    bri = float(brier_score_loss(y, proba))
    plt.figure(figsize=(4.8,4.8))
    plt.plot([0,1],[0,1],'--',label="Perfect")
    plt.plot(pp, pt, 'o-', label=f"{title} (ECE={ece:.3f}, Brier={bri:.3f})")
    plt.xlabel("Predicted probability")
    plt.ylabel("Empirical frequency")
    plt.title("Reliability (Calibration)")
    plt.legend()
    plt.tight_layout()
    plt.show()
    return ece, bri

print("\n[ROC/PR] Val: BNN only")
plot_roc_pr_single(yval, p_val, label="BNN", prefix="Val")
print("\n[ROC/PR] Test: BNN only")
plot_roc_pr_single(ytes, p_tst, label="BNN", prefix="Test")
print("\n[Calibration] Val: BNN only")
_ = reliability_plot(yval, p_val, "BNN (Val)")
print("\n[Calibration] Test: BNN only")
_ = reliability_plot(ytes, p_tst, "BNN (Test)")

def _get_arrays_from_globals():
    """Fetch BNN posterior means/variances for VAL/TEST from existing variables."""
    y_val_arr = np.asarray(yval).astype(int)
    y_tst_arr = np.asarray(ytes).astype(int)
    p_val_mean = np.asarray(p_val).ravel()
    p_tst_mean = np.asarray(p_tst).ravel()

    def _var_from_predictive(pred_dict):
        if isinstance(pred_dict, dict) and "p" in pred_dict:
            s = np.asarray(pred_dict["p"])
            return s.var(axis=0)
        return np.zeros_like(p_val_mean)

    var_val = _var_from_predictive(pred_val)
    var_tst = _var_from_predictive(pred_tst)

    return (y_val_arr, p_val_mean, var_val), (y_tst_arr, p_tst_mean, var_tst)

def binary_entropy(p, eps=1e-12):
    p = np.clip(p, eps, 1.0 - eps)
    return -(p*np.log(p) + (1-p)*np.log(1-p))

def ece_quantile(y_true, proba, n_bins=10):
    """ECE with quantile bins."""
    y_true = np.asarray(y_true).astype(int).ravel()
    proba  = np.asarray(proba).ravel()
    edges = np.unique(np.quantile(proba, np.linspace(0, 1, n_bins + 1)))
    edges[0], edges[-1] = 0.0, 1.0
    bin_idx = np.minimum(np.searchsorted(edges, proba, side="right") - 1, len(edges) - 2)
    bin_idx = np.clip(bin_idx, 0, len(edges) - 2)

    bin_pred, bin_true, bin_cnt = [], [], []
    ece_num, N = 0.0, len(proba)
    for b in range(len(edges) - 1):
        m = (bin_idx == b)
        cnt = int(m.sum())
        if cnt == 0:
            continue
        p_mean = float(proba[m].mean())
        t_rate = float(y_true[m].mean())
        bin_pred.append(p_mean)
        bin_true.append(t_rate)
        bin_cnt.append(cnt)
        ece_num += abs(p_mean - t_rate) * cnt

    ece = ece_num / max(N, 1)
    return ece, np.array(bin_pred), np.array(bin_true), np.array(bin_cnt), bin_idx, edges

(val_y, val_p, val_var), (tst_y, tst_p, tst_var) = _get_arrays_from_globals()

thr_ba = float(thr_bacc)

def balanced_accuracy(y_true, y_pred):
    y_true = np.asarray(y_true).astype(int)
    y_pred = np.asarray(y_pred).astype(int)
    tp = np.sum((y_pred==1) & (y_true==1))
    fn = np.sum((y_pred==0) & (y_true==1))
    tn = np.sum((y_pred==0) & (y_true==0))
    fp = np.sum((y_pred==1) & (y_true==0))
    tpr = tp / (tp + fn + 1e-12)
    tnr = tn / (tn + fp + 1e-12)
    return 0.5 * (tpr + tnr)

plt.figure(figsize=(11, 4))
plt.subplot(1, 2, 1)
plt.hist(val_p[val_y == 0], bins=50, alpha=0.6, label="Normal (y=0)", density=True)
plt.hist(val_p[val_y == 1], bins=50, alpha=0.6, label="Attack (y=1)",  density=True)
plt.title("Validation: Posterior Mean (p̄)")
plt.xlabel("Posterior mean p̄"); plt.ylabel("Density")
plt.legend()
plt.subplot(1, 2, 2)
plt.hist(tst_p[tst_y == 0], bins=50, alpha=0.6, label="Normal (y=0)", density=True)
plt.hist(tst_p[tst_y == 1], bins=50, alpha=0.6, label="Attack (y=1)",  density=True)
plt.title("Test: Posterior Mean (p̄)")
plt.xlabel("Posterior mean p̄"); plt.ylabel("Density")
plt.legend()
plt.tight_layout(); plt.show()

plt.figure(figsize=(11, 4))
plt.subplot(1, 2, 1)
plt.scatter(val_p[val_y == 0], val_var[val_y == 0], s=6, alpha=0.5, label="Normal (y=0)")
plt.scatter(val_p[val_y == 1], val_var[val_y == 1], s=6, alpha=0.5, label="Attack (y=1)")
plt.title("Validation: Mean vs Variance"); plt.xlabel("p̄"); plt.ylabel("Var[p]")
plt.legend(loc="upper center")
plt.subplot(1, 2, 2)
plt.scatter(tst_p[tst_y == 0], tst_var[tst_y == 0], s=6, alpha=0.5, label="Normal (y=0)")
plt.scatter(tst_p[tst_y == 1], tst_var[tst_y == 1], s=6, alpha=0.5, label="Attack (y=1)")
plt.title("Test: Mean vs Variance"); plt.xlabel("p̄"); plt.ylabel("Var[p]")
plt.legend(loc="upper center")
plt.tight_layout(); plt.show()

def reliability_with_variance(y_true, proba, var, title, n_bins=10):
    ece, bp, bt, bc, bin_idx, edges = ece_quantile(y_true, proba, n_bins=n_bins)
    var_bins = []
    for b in range(len(edges) - 1):
        m = (bin_idx == b)
        var_bins.append(float(var[m].mean()) if m.sum() else np.nan)
    var_bins = np.array(var_bins)
    size = 100 * (bc / bc.max())
    cmap = plt.cm.viridis

    plt.figure(figsize=(5.2, 5.2))
    plt.plot([0, 1], [0, 1], '--', color='gray', alpha=0.6, label="Perfect")
    sc = plt.scatter(bp, bt, s=size, c=var_bins, cmap=cmap, edgecolor='k', alpha=0.9)
    plt.xlabel("Predicted probability (bin mean)")
    plt.ylabel("Empirical frequency (bin rate)")
    plt.title(f"{title}\nECE={ece:.3f}")
    cbar = plt.colorbar(sc); cbar.set_label("Mean posterior variance per bin")
    for i in range(len(bp)):
        plt.text(bp[i], bt[i], str(int(bc[i])), fontsize=8, ha="center", va="bottom")
    plt.tight_layout(); plt.show()

reliability_with_variance(val_y, val_p, val_var, "Validation: Reliability & Uncertainty", n_bins=10)
reliability_with_variance(tst_y, tst_p, tst_var, "Test: Reliability & Uncertainty", n_bins=10)

def entropy_correctness_panels(y, p, thr, split_name, thr_name):
    H = binary_entropy(p)
    pred = (p >= thr).astype(int)
    correct = (pred == y)

    plt.figure(figsize=(11, 4))
    plt.subplot(1, 2, 1)
    plt.hist(H[correct],     bins=50, alpha=0.7, label="Correct",   density=True)
    plt.hist(H[~correct],    bins=50, alpha=0.7, label="Incorrect", density=True)
    plt.title(f"{split_name} ({thr_name}): Predictive Entropy")
    plt.xlabel("Entropy H(p̄)"); plt.ylabel("Density"); plt.legend()

    plt.subplot(1, 2, 2)
    plt.boxplot([H[correct], H[~correct]], labels=["Correct", "Incorrect"], showmeans=True)
    plt.title(f"{split_name} ({thr_name}): Entropy by correctness")
    plt.ylabel("Entropy H(p̄)")
    plt.tight_layout(); plt.show()

entropy_correctness_panels(val_y, val_p, thr_f1, "Validation", "F1-threshold")
entropy_correctness_panels(tst_y, tst_p, thr_f1, "Test",       "F1-threshold")
entropy_correctness_panels(val_y, val_p, thr_ba, "Validation", "BAcc-threshold")
entropy_correctness_panels(tst_y, tst_p, thr_ba, "Test",       "BAcc-threshold")

def coverage_curves(y, p, uncert, thr, metric="F1", n_pts=25):
    """
    Drop the most-uncertain q fraction (q up to 0.9), evaluate metric at fixed threshold.
    metric in {"F1","BAcc"}
    """
    m = y.shape[0]
    qs = np.linspace(0.0, 0.9, n_pts)
    coverages, vals = [], []
    for q in qs:
        tau = np.quantile(uncert, 1.0 - q)
        keep = (uncert <= tau)
        y_k, p_k = y[keep], p[keep]
        if y_k.size < 50:
            break
        coverages.append(y_k.size / m)
        pred_k = (p_k >= thr).astype(int)
        if metric == "F1":
            vals.append(f1_score(y_k, pred_k))
        elif metric == "BAcc":
            vals.append(balanced_accuracy(y_k, pred_k))
        else:
            raise ValueError("metric must be 'F1' or 'BAcc'")
    return np.array(coverages), np.array(vals)

u_val = val_var
u_tst = tst_var

for split, (y_, p_, u_) in [("Validation", (val_y, val_p, u_val)), ("Test", (tst_y, tst_p, u_tst))]:
    cov, f1s = coverage_curves(y_, p_, u_, thr_f1, metric="F1", n_pts=25)
    plt.figure(figsize=(5.8, 4.3))
    plt.plot(cov, f1s, marker='o')
    plt.gca().invert_xaxis()
    plt.xlabel("Coverage (kept ratio)"); plt.ylabel("F1 at fixed threshold")
    plt.title(f"{split}: Coverage vs F1 (F1-threshold)")
    plt.tight_layout(); plt.show()

for split, (y_, p_, u_) in [("Validation", (val_y, val_p, u_val)), ("Test", (tst_y, tst_p, u_tst))]:
    cov, baccs = coverage_curves(y_, p_, u_, thr_ba, metric="BAcc", n_pts=25)
    plt.figure(figsize=(5.8, 4.3))
    plt.plot(cov, baccs, marker='o')
    plt.gca().invert_xaxis()
    plt.xlabel("Coverage (kept ratio)"); plt.ylabel("Balanced Accuracy at fixed threshold")
    plt.title(f"{split}: Coverage vs BAcc (BAcc-threshold)")
    plt.tight_layout(); plt.show()

required = ["Xtr","Xval","Xtst","ytr","yval","ytes","p_val","p_tst","thr_f1","thr_bacc"]
missing = [v for v in required if v not in globals()]
if missing:
    raise RuntimeError(f"Missing variables from BNN pipeline: {missing}. Please run your BNN code first to define them.")

Xtr = np.asarray(Xtr, dtype=np.float32)
Xval = np.asarray(Xval, dtype=np.float32)
Xtst = np.asarray(Xtst, dtype=np.float32)
ytr  = np.asarray(ytr,  dtype=int)
yval = np.asarray(yval, dtype=int)
ytes = np.asarray(ytes, dtype=int)
p_val = np.asarray(p_val).ravel()
p_tst = np.asarray(p_tst).ravel()

def metrics_from_pred(y_true, y_pred):
    return dict(
        Acc = accuracy_score(y_true, y_pred),
        F1  = f1_score(y_true, y_pred),
        Precision = precision_score(y_true, y_pred, zero_division=0),
        Recall    = recall_score(y_true, y_pred, zero_division=0),
    )

def apply_minmax_clip(x, lo, hi):
    x = np.asarray(x, dtype=float)
    if hi <= lo:
        return np.zeros_like(x)
    return np.clip((x - lo) / (hi - lo), 0.0, 1.0)

thr_f1   = float(thr_f1)
thr_bacc = float(thr_bacc)

bnn_pred_val_f1    = (p_val >= thr_f1).astype(int)
bnn_pred_test_f1   = (p_tst >= thr_f1).astype(int)
bnn_pred_val_bacc  = (p_val >= thr_bacc).astype(int)
bnn_pred_test_bacc = (p_tst >= thr_bacc).astype(int)

metrics_val = {}
metrics_tst = {}
prob_val = {"BNN": p_val.copy()}
prob_tst = {"BNN": p_tst.copy()}

metrics_val["BNN(F1)"]   = metrics_from_pred(yval, bnn_pred_val_f1)
metrics_tst["BNN(F1)"]   = metrics_from_pred(ytes, bnn_pred_test_f1)
metrics_val["BNN(BAcc)"] = metrics_from_pred(yval, bnn_pred_val_bacc)
metrics_tst["BNN(BAcc)"] = metrics_from_pred(ytes, bnn_pred_test_bacc)

print("[BNN] thresholds:", f"F1-opt={thr_f1:.4f}, BAcc-opt={thr_bacc:.4f}")

print("\n[LOF] Training on normal-only samples (y==0)...")
normal_mask_tr = (ytr == 0)
Xtr_norm = Xtr[normal_mask_tr]

lof = LocalOutlierFactor(
    n_neighbors=35, contamination=0.1, novelty=True, n_jobs=-1
)
lof.fit(Xtr_norm)

score_val_lof = -lof.decision_function(Xval)
score_tst_lof = -lof.decision_function(Xtst)

lof_lo, lof_hi = float(np.min(score_val_lof)), float(np.max(score_val_lof))
prob_val["LOF"] = apply_minmax_clip(score_val_lof, lof_lo, lof_hi)
prob_tst["LOF"] = apply_minmax_clip(score_tst_lof, lof_lo, lof_hi)

thr_lof = best_f1_threshold(yval, prob_val["LOF"])
pred_val_lof = (prob_val["LOF"] >= thr_lof).astype(int)
pred_tst_lof = (prob_tst["LOF"] >= thr_lof).astype(int)

metrics_val["LOF"] = metrics_from_pred(yval, pred_val_lof)
metrics_tst["LOF"] = metrics_from_pred(ytes, pred_tst_lof)
print(f"[LOF] Val F1-opt threshold={thr_lof:.6f}")

def auto_eps_knn(X, k=5, q=0.95, max_sample=20000, random_state=42):
    rng = np.random.default_rng(random_state)
    if X.shape[0] > max_sample:
        idx = rng.choice(X.shape[0], size=max_sample, replace=False)
        Xs = X[idx]
    else:
        Xs = X
    nn = NearestNeighbors(n_neighbors=k+1, n_jobs=-1).fit(Xs)
    dists, _ = nn.kneighbors(Xs)
    kth = dists[:, -1]
    return float(np.quantile(kth, q))

print("\n[DBSCAN] Auto-selecting eps...")
eps = auto_eps_knn(Xtr, k=5, q=0.95)
print(f"[DBSCAN] Using eps={eps:.6f}, min_samples=5")
dbs = DBSCAN(eps=eps, min_samples=5, n_jobs=-1)
dbs.fit(Xtr)

labels_tr = dbs.labels_
core_mask = np.zeros_like(labels_tr, dtype=bool)
if hasattr(dbs, "core_sample_indices_"):
    core_mask[dbs.core_sample_indices_] = True

cluster_map = {}
cluster_prob = {}
for lab in np.unique(labels_tr):
    idx = np.where(labels_tr == lab)[0]
    if (lab == -1) or (len(idx) == 0):
        cluster_map[lab]  = 1
        cluster_prob[lab] = 1.0
    else:
        y_mean = float(np.mean(ytr[idx]))
        cluster_map[lab]  = 1 if y_mean >= 0.5 else 0
        cluster_prob[lab] = y_mean

if core_mask.sum() == 0:
    X_core = Xtr
    lab_core = labels_tr
else:
    X_core = Xtr[core_mask]
    lab_core = labels_tr[core_mask]

nn_core = NearestNeighbors(n_neighbors=1, n_jobs=-1).fit(X_core)

def dbscan_predict_and_prob(X):
    _, idx = nn_core.kneighbors(X, return_distance=True)
    nearest_labels = lab_core[idx.ravel()]
    y_hat = np.array([cluster_map.get(int(lab), 1) for lab in nearest_labels], dtype=int)
    p_hat = np.array([cluster_prob.get(int(lab), 1.0) for lab in nearest_labels], dtype=float)
    return y_hat, p_hat

pred_val_dbs, prob_val_dbs = dbscan_predict_and_prob(Xval)
pred_tst_dbs, prob_tst_dbs = dbscan_predict_and_prob(Xtst)

metrics_val["DBSCAN"] = metrics_from_pred(yval, pred_val_dbs)
metrics_tst["DBSCAN"] = metrics_from_pred(ytes, pred_tst_dbs)

prob_val["DBSCAN"] = prob_val_dbs
prob_tst["DBSCAN"] = prob_tst_dbs

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

class AE(nn.Module):
    def __init__(self, d_in, bottleneck=16):
        super().__init__()
        self.enc = nn.Sequential(
            nn.Linear(d_in, 128), nn.ReLU(),
            nn.Linear(128, 64), nn.ReLU(),
            nn.Linear(64, bottleneck)
        )
        self.dec = nn.Sequential(
            nn.Linear(bottleneck, 64), nn.ReLU(),
            nn.Linear(64, 128), nn.ReLU(),
            nn.Linear(128, d_in)
        )
    def forward(self, x):
        z = self.enc(x)
        xhat = self.dec(z)
        return xhat

def train_ae(X_train, epochs=20, batch_size=1024, lr=1e-3, max_train=60000):
    Xn = Xtr[ytr == 0]
    if Xn.shape[0] > max_train:
        rng = np.random.default_rng(42)
        idx = rng.choice(Xn.shape[0], size=max_train, replace=False)
        Xn = Xn[idx]
    d = Xn.shape[1]
    model = AE(d_in=d, bottleneck=16).to(device)
    opt = torch.optim.Adam(model.parameters(), lr=lr)
    ds = TensorDataset(torch.from_numpy(Xn))
    dl = DataLoader(ds, batch_size=batch_size, shuffle=True, drop_last=False)
    loss_fn = nn.MSELoss(reduction="mean")

    model.train()
    for ep in range(epochs):
        run_loss = 0.0
        for (xb,) in dl:
            xb = xb.to(device)
            opt.zero_grad(set_to_none=True)
            xhat = model(xb)
            loss = loss_fn(xhat, xb)
            loss.backward()
            opt.step()
            run_loss += float(loss.item()) * xb.size(0)
        run_loss /= Xn.shape[0]
        if (ep+1) % 5 == 0:
            print(f"[AE] epoch {ep+1}/{epochs}  loss={run_loss:.6f}")
    return model

def recon_error(model, X):
    model.eval()
    with torch.no_grad():
        xb = torch.from_numpy(np.asarray(X, dtype=np.float32)).to(device)
        xhat = model(xb)
        err = torch.mean((xhat - xb)**2, dim=1).cpu().numpy()
    return err

print("\n[AE] Training...")
ae = train_ae(Xtr, epochs=20, batch_size=1024, lr=1e-3, max_train=60000)
err_val_ae = recon_error(ae, Xval)
err_tst_ae = recon_error(ae, Xtst)

ae_lo, ae_hi = float(np.min(err_val_ae)), float(np.max(err_val_ae))
prob_val["AE"]  = apply_minmax_clip(err_val_ae,  ae_lo, ae_hi)
prob_tst["AE"]  = apply_minmax_clip(err_tst_ae,  ae_lo, ae_hi)

thr_ae = best_f1_threshold(yval, prob_val["AE"])
pred_val_ae = (prob_val["AE"] >= thr_ae).astype(int)
pred_tst_ae = (prob_tst["AE"] >= thr_ae).astype(int)

metrics_val["AE"] = metrics_from_pred(yval, pred_val_ae)
metrics_tst["AE"] = metrics_from_pred(ytes, pred_tst_ae)
print(f"[AE] Val F1-opt threshold={thr_ae:.6f}")

class VAE(nn.Module):
    def __init__(self, d_in, z_dim=8):
        super().__init__()
        self.enc = nn.Sequential(
            nn.Linear(d_in, 128), nn.ReLU(),
            nn.Linear(128, 64), nn.ReLU(),
        )
        self.mu = nn.Linear(64, z_dim)
        self.logvar = nn.Linear(64, z_dim)
        self.dec = nn.Sequential(
            nn.Linear(z_dim, 64), nn.ReLU(),
            nn.Linear(64, 128), nn.ReLU(),
            nn.Linear(128, d_in)
        )
    def encode(self, x):
        h = self.enc(x)
        return self.mu(h), self.logvar(h)
    def reparam(self, mu, logvar):
        std = torch.exp(0.5 * logvar)
        eps = torch.randn_like(std)
        return mu + eps * std
    def forward(self, x):
        mu, logvar = self.encode(x)
        z = self.reparam(mu, logvar)
        xhat = self.dec(z)
        return xhat, mu, logvar

def vae_loss(x, xhat, mu, logvar):
    recon = nn.MSELoss(reduction="mean")(xhat, x)
    kl = -0.5 * torch.mean(1 + logvar - mu.pow(2) - logvar.exp())
    return recon + kl, recon.detach(), kl.detach()

def train_vae(X_train, epochs=25, batch_size=1024, lr=1e-3, max_train=40000):
    Xn = Xtr[ytr == 0]
    if Xn.shape[0] > max_train:
        rng = np.random.default_rng(42)
        idx = rng.choice(Xn.shape[0], size=max_train, replace=False)
        Xn = Xn[idx]
    d = Xn.shape[1]
    model = VAE(d_in=d, z_dim=8).to(device)
    opt = torch.optim.Adam(model.parameters(), lr=lr)
    ds = TensorDataset(torch.from_numpy(Xn))
    dl = DataLoader(ds, batch_size=batch_size, shuffle=True, drop_last=False)

    model.train()
    for ep in range(epochs):
        run_loss = 0.0
        for (xb,) in dl:
            xb = xb.to(device)
            opt.zero_grad(set_to_none=True)
            xhat, mu, logvar = model(xb)
            loss, _, _ = vae_loss(xb, xhat, mu, logvar)
            loss.backward()
            nn.utils.clip_grad_norm_(model.parameters(), max_norm=5.0)
            opt.step()
            run_loss += float(loss.item()) * xb.size(0)
        run_loss /= Xn.shape[0]
        if (ep+1) % 5 == 0:
            print(f"[VAE] epoch {ep+1}/{epochs}  loss={run_loss:.6f}")
    return model

def vae_score(model, X, num_samples=1):
    """Anomaly score proxy -> convert to [0,1] prob later."""
    model.eval()
    with torch.no_grad():
        xb = torch.from_numpy(np.asarray(X, dtype=np.float32)).to(device)
        total = []
        for _ in range(num_samples):
            xhat, mu, logvar = model(xb)
            err = torch.mean((xhat - xb)**2, dim=1) + 0.5 * torch.sum(
                mu**2 + torch.exp(logvar) - 1 - logvar, dim=1
            )
            total.append(err.cpu().numpy())
        s = np.mean(np.stack(total, axis=0), axis=0)
    return s

print("\n[VAE] Training...")
vae = train_vae(Xtr, epochs=25, batch_size=1024, lr=1e-3, max_train=40000)
score_val_vae = vae_score(vae, Xval, num_samples=1)
score_tst_vae = vae_score(vae, Xtst, num_samples=1)

vae_lo, vae_hi = float(np.min(score_val_vae)), float(np.max(score_val_vae))
prob_val["VAE"] = apply_minmax_clip(score_val_vae, vae_lo, vae_hi)
prob_tst["VAE"] = apply_minmax_clip(score_tst_vae, vae_lo, vae_hi)

thr_vae = best_f1_threshold(yval, prob_val["VAE"])
pred_val_vae = (prob_val["VAE"] >= thr_vae).astype(int)
pred_tst_vae = (prob_tst["VAE"] >= thr_vae).astype(int)

metrics_val["VAE"] = metrics_from_pred(yval, pred_val_vae)
metrics_tst["VAE"] = metrics_from_pred(ytes, pred_tst_vae)
print(f"[VAE] Val F1-opt threshold={thr_vae:.6f}")

def clamp01_dict(d):
    for k in list(d.keys()):
        d[k] = np.clip(np.asarray(d[k], dtype=float), 0.0, 1.0)

clamp01_dict(prob_val)
clamp01_dict(prob_tst)

order = ["BNN(F1)", "BNN(BAcc)", "LOF", "DBSCAN", "AE", "VAE"]
cols_report  = ["Acc","F1","Precision","Recall","ECE","Brier"]

metrics_val = {k: dict(v) for k,v in metrics_val.items()}
metrics_tst = {k: dict(v) for k,v in metrics_tst.items()}

prob_map = {
    "BNN(F1)": "BNN",
    "BNN(BAcc)": "BNN",
    "LOF": "LOF",
    "DBSCAN": "DBSCAN",
    "AE": "AE",
    "VAE": "VAE",
}

for name in order:
    pk = prob_map[name]
    pv = prob_val[pk]
    pt = prob_tst[pk]
    metrics_val[name]["ECE"]   = ece_quantile(yval, pv, n_bins=10)[0]
    metrics_val[name]["Brier"] = brier_score_loss(yval, pv)
    metrics_tst[name]["ECE"]   = ece_quantile(ytes, pt, n_bins=10)[0]
    metrics_tst[name]["Brier"] = brier_score_loss(ytes, pt)

df_val = pd.DataFrame({k: metrics_val[k] for k in order}).T[cols_report]
df_tst = pd.DataFrame({k: metrics_tst[k] for k in order}).T[cols_report]

print("\n[Validation] metrics (unweighted):")
display(df_val.style.format("{:.4f}"))
print("\n[Test] metrics (unweighted):")
display(df_tst.style.format("{:.4f}"))

def grouped_bars(metrics_dict, title):
    models = order
    metrics = ["Acc","F1","Precision","Recall"]
    M = len(models)
    K = len(metrics)

    width = 0.18
    x = np.arange(M)

    plt.figure(figsize=(12, 6))
    for i, m in enumerate(metrics):
        vals = [metrics_dict[model][m] for model in models]
        plt.bar(x + (i - (K-1)/2)*width, vals, width, label=m)
        for j, v in enumerate(vals):
            plt.text(x[j] + (i - (K-1)/2)*width, v + 0.01, f"{v:.3f}",
                     ha="center", va="bottom", fontsize=8)

    plt.xticks(x, models, rotation=20, ha="right")
    plt.ylim(0, 1.05)
    plt.ylabel("Score")
    plt.title(title)
    plt.legend()
    plt.tight_layout()
    plt.show()

grouped_bars(metrics_val, "Validation — Accuracy / F1 / Precision / Recall (unweighted)")
grouped_bars(metrics_tst, "Test — Accuracy / F1 / Precision / Recall (unweighted)")

HAS_XGB = XGBClassifier is not None
rs = check_random_state(42)

feat_names_red = np.array([b for b in base_names if b in set(selected)])
if feat_names_red.ndim == 0:
    feat_names_red = np.array([str(feat_names_red)])

if HAS_XGB:
    clf = XGBClassifier(
        n_estimators=300, max_depth=4, learning_rate=0.05,
        subsample=0.9, colsample_bytree=0.9,
        reg_alpha=1e-3, reg_lambda=1e-2,
        random_state=42, n_jobs=-1, tree_method="hist"
    )
else:
    from sklearn.linear_model import LogisticRegression
    clf = LogisticRegression(
        solver="saga", max_iter=2000, n_jobs=-1, class_weight="balanced", random_state=42
    )

clf.fit(Xtr, ytr)

thr_bnn = float(thr_bacc)

K_BG = 64
background = shap.kmeans(Xtr, K_BG)
print(f"[SHAP-Proxy] k-means background K={K_BG}")

p_sur_tst = clf.predict_proba(Xtst)[:, 1]
pred_sur_tst = (p_sur_tst >= thr_bnn).astype(int)
mis_mask = pred_sur_tst != ytes

idx_mis = np.where(mis_mask)[0]
idx_cor = np.where(~mis_mask)[0]
take_mis = min(250, len(idx_mis))
take_cor = min(350, len(idx_cor))
idx_exp = np.r_[rs.choice(idx_mis, size=take_mis, replace=False) if take_mis>0 else [],
                rs.choice(idx_cor, size=take_cor, replace=False)]
rs.shuffle(idx_exp)

X_exp = Xtst[idx_exp]
y_exp = ytes[idx_exp]
print(f"[Explain-Proxy] n={len(idx_exp)} (BNN-mis={take_mis}, correct={take_cor})")

explainer = shap.KernelExplainer(lambda X: clf.predict_proba(X)[:, 1], background)
sv_proxy = explainer.shap_values(X_exp, nsamples="auto")
expected_value_proxy = np.atleast_1d(explainer.expected_value).mean()
sv_proxy = np.asarray(sv_proxy)
print("[SHAP-Proxy] values shape:", sv_proxy.shape)

plt.figure(figsize=(6, 7))
shap.summary_plot(sv_proxy, X_exp, feature_names=feat_names_red, show=False, plot_type="dot")
plt.title("SHAP summary (beeswarm) — Test subset (Proxy)")
plt.tight_layout(); plt.show()

plt.figure(figsize=(6, 8))
shap.summary_plot(sv_proxy, X_exp, feature_names=feat_names_red, show=False, plot_type="bar")
plt.title("Global feature importance by SHAP — Test subset (Proxy)")
plt.tight_layout(); plt.show()

def to_base(col: str) -> str:
    return base_of(col)

feat_base = np.array([to_base(c) for c in feat_names_red])
abs_mean_shap_ohe = np.mean(np.abs(sv_proxy), axis=0)
df_shap_base = (
    pd.DataFrame({"base": feat_base, "abs_mean_shap_ohe": abs_mean_shap_ohe})
    .groupby("base", as_index=False)["abs_mean_shap_ohe"].sum()
    .rename(columns={"abs_mean_shap_ohe": "abs_mean_shap_base"})
    .sort_values("abs_mean_shap_base", ascending=False)
    .reset_index(drop=True)
)
print("[SHAP-Proxy] Aggregated by BASE feature (sum |SHAP|):")
display(df_shap_base.head(20))

def perm_importance_brier(model, X_val, y_val, feature_names, n_repeats=5, random_state=42):
    rng = np.random.RandomState(random_state)
    base = brier_score_loss(y_val, model.predict_proba(X_val)[:,1])
    rows = []
    for j, f in enumerate(feature_names):
        deltas = []
        for _ in range(n_repeats):
            Xp = X_val.copy()
            rng.shuffle(Xp[:, j])
            p = model.predict_proba(Xp)[:,1]
            deltas.append(brier_score_loss(y_val, p) - base)
        rows.append((f, float(np.mean(deltas))))
    df = (
        pd.DataFrame(rows, columns=["feature","perm_imp_brier"])
        .sort_values("perm_imp_brier", ascending=False)
    )
    return df

df_perm_ohe = perm_importance_brier(clf, Xval, yval, feat_names_red, n_repeats=5)
df_perm_ohe["base"] = [to_base(c) for c in df_perm_ohe["feature"]]
df_perm_base = (
    df_perm_ohe.groupby("base", as_index=False)["perm_imp_brier"].sum()
    .rename(columns={"perm_imp_brier": "perm_imp_brier_base"})
    .sort_values("perm_imp_brier_base", ascending=False)
    .reset_index(drop=True)
)
print("[Permutation-Proxy] (Brier) by BASE feature:")
display(df_perm_base.head(20))

def bnn_predict_proba_mean(X_np, n_samples=64, seed=2025):
    X_jax = jnp.array(X_np.astype(np.float32))
    pred  = Predictive(model=model_uncertain, guide=guide,
                       params=params, num_samples=n_samples)
    out   = pred(rng_key=random.PRNGKey(seed), x=X_jax, y=None, hidden_dim=H)
    p     = np.asarray(out["p"]).mean(axis=0)
    return p

mis_idx_all = np.where(mis_mask)[0]
if len(mis_idx_all) < 2:
    uncert = np.abs(p_sur_tst - 0.5)
    mis_idx_all = np.argsort(uncert)[:2]

idx_local = mis_idx_all[:2]
print(f"[Local-BNN] explain idx={idx_local.tolist()} (y={ytes[idx_local].tolist()})")

K_BG_LOCAL = 25
background_local = shap.kmeans(Xtr, K_BG_LOCAL)
print(f"[Local-BNN] k-means background K={K_BG_LOCAL}")

try:
    explainer_local = shap.SamplingExplainer(
        model=lambda X: bnn_predict_proba_mean(np.asarray(X), n_samples=64),
        data=background_local
    )
    sv_local = explainer_local.shap_values(Xtst[idx_local], max_evals=5000)
except Exception as e:
    print(f"[Local-BNN] SamplingExplainer not available -> fallback KernelExplainer ({e})")
    explainer_local = shap.KernelExplainer(
        model=lambda X: bnn_predict_proba_mean(np.asarray(X), n_samples=64),
        data=background_local
    )
    sv_local = explainer_local.shap_values(Xtst[idx_local], nsamples="auto")

sv_local = np.asarray(sv_local)
print(f"[Local-BNN] SHAP values shape: {sv_local.shape}")

for k, idx0 in enumerate(idx_local, 1):
    plt.figure(figsize=(10, 3.2))
    shap.decision_plot(
        base_value=np.atleast_1d(explainer_local.expected_value).mean(),
        shap_values=sv_local[[k-1]],
        features=Xtst[[idx0]],
        feature_names=feat_names_red,
        link="logit",
        show=False
    )
    plt.title(f"BNN (local) — SHAP decision plot — test idx={idx0}")
    plt.tight_layout()
    plt.show()
    plt.close()

    fig = shap.force_plot(
        base_value=np.atleast_1d(explainer_local.expected_value).mean(),
        shap_values=sv_local[k-1],
        features=Xtst[idx0],
        feature_names=feat_names_red,
        matplotlib=True,
        show=False
    )
    plt.gcf().suptitle(f"BNN (local) — SHAP force plot — test idx={idx0}", y=0.98)
    plt.tight_layout()
    plt.show()
    plt.close()
```
